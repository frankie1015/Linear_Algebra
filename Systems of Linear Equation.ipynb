```python
from sympy import symbols, Eq, solve

# Define the variables and the parameter
x1, x2, x3, t = symbols('x1 x2 x3 t')

# Given system of equations
equation1 = Eq(x1 - 3*x2 - 2*x3, 0)
equation2 = Eq(-x1 + 2*x2 + x3, 0)
equation3 = Eq(2*x1 + 4*x2 + 6*x3, 0)

# Express x1 in terms of x2 and x3 using the second equation
x1_expr = solve(equation2, x1)[0]

# Substitute x1 back into the other equations
equation1 = equation1.subs(x1, x1_expr)
equation3 = equation3.subs(x1, x1_expr)

# Solve the system of equations
solution = solve((equation1, equation2, equation3), (x2, x3))

# Print the solution
print(f"x1 = {x1_expr}")
print(f"x2 = {solution[x2]}")
print(f"x3 = {solution[x3]}")
print(f"t is a parameter")

```

```python
from sympy import symbols, Eq, solve

# Define the variables
w, x, y, z = symbols('w x y z')

# Given system of equations
equation1 = Eq(w + x + 2*y + z, 1)
equation2 = Eq(w - x - y + z, 0)
equation3 = Eq(x + y, 1)
equation4 = Eq(w + x + z, 2)

# Solve the system of equations
solutions = solve((equation1, equation2, equation3, equation4), (w, x, y, z))

# Check if there are solutions
if solutions:
    # Print all solutions
    for sol in solutions:
        print("Solution:")
        for var in [w, x, y, z]:
            print(f"{var} = {sol[var]}")
        print()
else:
    print("No solution found.")

```
