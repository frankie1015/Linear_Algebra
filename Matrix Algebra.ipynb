Without the use of any python libraries or modules, develop a function that can perform matrix addition, given two numpy matrices.

```python
def matrix_addition(matrix1, matrix2):
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        raise ValueError("Matrices must have the same dimensions for addition.")
    result_matrix = [[0] * len(matrix1[0]) for _ in range(len(matrix1))]

    for i in range(len(matrix1)):
        for j in range(len(matrix1[0])):
            result_matrix[i][j] = matrix1[i][j] + matrix2[i][j]

    return result_matrix
matrix_a = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

matrix_b = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]

result = matrix_addition(matrix_a, matrix_b)
for row in result:
    print(row)
```

Without the use of any python libraries or modules, develop a function that can multiply two numpy matrices.

```python
def matrix_multiplication(matrix1, matrix2):
    if len(matrix1[0]) != len(matrix2):
        raise ValueError("Number of columns in the first matrix must be equal to the number of rows in the second matrix.")
    result_matrix = [[0] * len(matrix2[0]) for _ in range(len(matrix1))]

    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                result_matrix[i][j] += matrix1[i][k] * matrix2[k][j]

    return result_matrix
matrix_a = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

matrix_b = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]

result = matrix_multiplication(matrix_a, matrix_b)
for row in result:
    print(row)
```

The rule of distributivity states that given two matrices ùê¥ and ùêµ and a scalar, ùëò, then ùëò(ùê¥+ùêµ)=ùëòùê¥+ùëòùêµ. Instead of writing a proof mathematically, develop two codes for ùëò(ùê¥+ùêµ) and ùëòùê¥+ùëòùêµ.

```python
def scalar_matrix_addition(scalar, matrix1, matrix2):
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        raise ValueError("Matrices must have the same dimensions for addition.")
    result_matrix = [[0] * len(matrix1[0]) for _ in range(len(matrix1))]

    for i in range(len(matrix1)):
        for j in range(len(matrix1[0])):
            result_matrix[i][j] = scalar * (matrix1[i][j] + matrix2[i][j])

    return result_matrix


def scalar_matrix_multiplication(scalar, matrix):
    result_matrix = [[0] * len(matrix[0]) for _ in range(len(matrix))]
    for i in range(len(matrix)):
        for j in range(len(matrix[0])):
            result_matrix[i][j] = scalar * matrix[i][j]

    return result_matrix
scalar = 2
matrix_a = [
    [1, 2],
    [3, 4]
]

matrix_b = [
    [5, 6],
    [7, 8]
]

# k(A + B)
result_addition = scalar_matrix_addition(scalar, matrix_a, matrix_b)

# kA + kB
result_multiplication = scalar_matrix_addition(scalar, matrix_a, matrix_b)

# Print results
print("k(A + B):")
for row in result_addition:
    print(row)

print("\nkA + kB:")
for row in result_multiplication:
    print(row)
```

Without using a python library or modules develop a function that can extract the diagonal of a numpy matrix.

```python
def extract_diagonal(matrix):
    if len(matrix) != len(matrix[0]):
        raise ValueError("Matrix must be square to extract the diagonal.")

    diagonal = [matrix[i][i] for i in range(len(matrix))]

    return diagonal

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

result = extract_diagonal(matrix)
print("Diagonal elements:", result)
```

Without using a python library or modules develop a function that can find a trace of a numpy matrix.

```python
def matrix_trace(matrix):
    if len(matrix) != len(matrix[0]):
        raise ValueError("Matrix must be square to find the trace.")
    trace_sum = sum(matrix[i][i] for i in range(len(matrix)))

    return trace_sum

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

result = matrix_trace(matrix)
print("Trace of the matrix:", result)
```
